```python
dict_table={}

for letter, i in zip(alphabet, range(6, 42, 7)):
    df = nation[i]
    print(letter , i)
    dict_table[f'Group {letter}'] = df
```

```python
with open('dict_table', 'wb') as output : 
    pickle.dump(dict_table, output)
```
