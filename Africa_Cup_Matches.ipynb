```python
import re

# Apply the transformation on the 'score' column
def format_score(score):
    # If the score contains a space (e.g., '2 1')
    if isinstance(score, str) and ' ' in score:
        return '-'.join(re.findall(r'\d+', score))
    # If the score is a single number with no space (e.g., '12')
    elif isinstance(score, str) and len(score) == 2 and score.isdigit():
        return score[0] + '-' + score[1]
    return score

# Apply the function to the 'score' column
df_historical_matches['score'] = df_historical_matches['score'].apply(format_score)
```

```python
df_historical_matches[['HomeGoals','AwayGoals']] = df_historical_matches['score'].str.split('-', expand = True)
```
